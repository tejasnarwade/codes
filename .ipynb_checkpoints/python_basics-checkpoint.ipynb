```python
print("------Operators in Python------")
'''

🔹 1. Arithmetic Operators
Operator	Name	Example	Result
+	Addition	5 + 2	7
-	Subtraction	5 - 2	3
*	Multiplication	5 * 2	10
/	Division	5 / 2	2.5
//	Floor Division	5 // 2	2
%	Modulus	5 % 2	1
**	Exponent	5 ** 2	25
🔹 2. Comparison (Relational) Operators
Operator	Meaning	Example
==	Equal to	a == b
!=	Not equal to	a != b
>	Greater than	a > b
<	Less than	a < b
>=	Greater than or equal to	a >= b
<=	Less than or equal to	a <= b
🔹 3. Assignment Operators
Operator	Example	Equivalent To
=	a = 5	—
+=	a += 2	a = a + 2
-=	a -= 2	a = a - 2
*=	a *= 2	a = a * 2
/=	a /= 2	a = a / 2
//=	a //= 2	a = a // 2
%=	a %= 2	a = a % 2
**=	a **= 2	a = a ** 2
&=	a &= 2	a = a & 2
`	=`	`a
^=	a ^= 2	a = a ^ 2
>>=	a >>= 2	a = a >> 2
<<=	a <<= 2	a = a << 2
🔹 4. Logical Operators
Operator	Meaning	Example
and	Logical AND	a and b
or	Logical OR	a or b
not	Logical NOT	not a
🔹 5. Bitwise Operators
Operator	Meaning	Example
&	AND	a & b
|	OR	a | b
^	XOR	a ^ b
~	NOT (invert)	~a
<<	Left shift	a << 2
>>	Right shift	a >> 2
🔹 6. Membership Operators
Operator	Meaning	Example
in	Present in sequence	x in list
not in	Not present	x not in list
🔹 7. Identity Operators
Operator	Meaning	Example
is	Same object	a is b
is not	Not same object	a is not b
🔹 8. Special Operators
Operator	Meaning	Example
:=	Walrus (Assignment expression)	if (n := len(arr)) > 5:

'''
print("----ARITHMETIC OPERATOR--------")
n1 = 10
n2 = 4
print(n1+n2) # adds numbers, output 14
print(n1-n2) # subtracts numbers, output 6
print(n1*n2) # multiplies numbers, output 40
print(n1/n2) # normal division (float), output 2.5
print(n1//n2) # floor division (integer part), output 2
print(n1%n2) # remainder after division, output 2
print(2**32) # power operation, output 4294967296

print("-------Comparison/RElationsal operator-----------")
# relational operator returns True or False
print(n1==n2) # checks equal, output False
print(n1!=n2) # checks not equal, output True
print(n1>n2)  # checks greater than, output True
print(n1<n2)  # checks less than, output False
print(n1<=n2) # checks less or equal, output False
print(n1>=n2) # checks greater or equal, output True

print("--------Assignment operator--------")

a = 10
print("Initial a =", a) # starting value 10

a += 5
print("a += 5 =", a) # adds 5, output 15

a -= 3
print("a -= 3 =", a) # subtracts 3, output 12

a *= 2
print("a *= 2 =", a) # multiply by 2, output 24

a /= 4
print("a /= 4 =", a) # divide by 4 (float), output 6.0

a //= 2
print("a //= 2 =", a) # floor divide, output 3.0

a %= 3
print("a %= 3 =", a) # remainder, output 0.0

a **= 2
print("a **= 2 =", a) # power, output 0.0

a = 5
a &= 3
print("a &= 3 =", a) # bitwise AND, output 1

a = 5
a |= 3
print("a |= 3 =", a) # bitwise OR, output 7

a = 5
a ^= 3
print("a ^= 3 =", a) # bitwise XOR, output 6

a = 8
a >>= 2
print("a >>= 2 =", a) # right shift, output 2

a = 8
a <<= 2
print("a <<= 2 =", a) # left shift, output 32


print("--------Logical Operators--------")

x = True
y = False

print("x and y =", x and y) # both must be True, output False
print("x or y =", x or y)   # at least one True, output True
print("not x =", not x)     # reverses value, output False


print("--------Bitwise Operators--------")

b1 = 5
b2 = 3

print("b1 & b2 =", b1 & b2) # bitwise AND, output 1
print("b1 | b2 =", b1 | b2) # bitwise OR, output 7
print("b1 ^ b2 =", b1 ^ b2) # bitwise XOR, output 6
print("~b1 =", ~b1)         # bitwise NOT, output -6
print("b1 << 1 =", b1 << 1) # left shift, output 10
print("b1 >> 1 =", b1 >> 1) # right shift, output 2


print("--------Membership Operators--------")

lst = [1, 2, 3, 4]

print("2 in lst =", 2 in lst) # checks if 2 exists, output True
print("10 not in lst =", 10 not in lst) # checks if 10 not exists, output True


print("--------Identity Operators--------")

m = [1, 2]
n = [1, 2]
p = m

print("m is n =", m is n) # different objects, output False
print("m is p =", m is p) # same object, output True
print("m is not n =", m is not n) # different objects, output True


print("--------Walrus Operator--------")
# walrus assigns value inside condition and returns it
if (length := len(lst)) > 2:
    print("Length is", length) # length is 4cx 
```

```python
#Functions in python 
print("-------functions in python---------")
'''
syntax :

def functionname (paramters):
    #code
    return expression


//calling a function

functionname(arguments)
'''
#types of function argument: default argument function, keyword argument
#postional argument, arbitary argument (variabel argument)

#types of function:
#anonymous function, pass by value , pass by reference

print(" deafult argument function")
#A default argument is a parameter that assumes a default value if a value is not provided in the function call for that argument.
def  myfunc(x,y=10):
    print(x,y)
myfunc(15)

print(" keyword argument")
#In keyword arguments, values are passed by explicitly specifying the parameter names, so the order doesn’t matter.
def student(fname,lname):
    print("full name is : ",fname,lname)
student(fname="tejas",lname="narwade")
student(lname="narwade",fname="tejas")

print("positional argument")
#In positional arguments, values are assigned to parameters based on their order in the function call.
def nameAge(name, age):
    print("Hi, I am", name)
    print("My age is ", age)

nameAge("Suraj", 27)

print("arbitrary arguments")
'''
In Python Arbitrary Keyword Arguments, *args and **kwargs can pass a variable number of arguments to a function using special symbols. There are two special symbols:

*args in Python (Non-Keyword Arguments)
**kwargs in Python (Keyword Arguments)

'''

def myFun(*args, **kwargs):
    print("Non-Keyword Arguments (*args):")
    for arg in args:
        print(arg)

    print("\nKeyword Arguments (**kwargs):")
    for key, value in kwargs.items():
        print(f"{key} == {value}")

# Function call with both types of arguments
myFun('Hey', 'Welcome', first='Geeks', mid='for', last='Geeks')

print("anonymous function : ")
def cube(x): return x*x*x   #normal ,, non anonymous function
#anonymous function using lambda
cubef  = lambda x : x*x*x
print(cubef(7))
'''
lambda function or anonymous function is a function without name 
used to write small functions as lamda function can have only one expression
no loops nothing just an expression 
'''
#rule of thumb lambda vs def ; if expression or code fit in one line then lambda can be used else use def normal function for each time
name = "pikachu"
mys = lambda name : name.replace('h','5').split("k")
print(mys(name))
```
